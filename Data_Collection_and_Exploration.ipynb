#Import Our Libraries

```python
import pandas as pd
import numpy as np
```

# Importing the data & Gaining Insights

```python
df = pd.read_csv("customer_churn_dataset-training-master.csv")
```

```python
df2 = pd.read_csv("customer_churn_dataset-testing-master.csv")
```

```python
df = pd.concat([df, df2], ignore_index = True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

#Distribution of Age

```python
sns.histplot(df['Age'], kde=True)
plt.show()
```

# Example: Correlation Heatmap

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

# Count of customers by Gender

```python
sns.countplot(x='Gender', data=df)
plt.title('Customer Count by Gender')
plt.show()
```

# Count of customers by Subscription Type

```python
sns.countplot(x='Subscription Type', data=df)
plt.title('Customer Count by Subscription Type')
plt.xticks(rotation=45)
plt.show()
```

# Count of customers by Contract Length

```python
sns.countplot(x='Contract Length', data=df)
plt.title('Customer Count by Contract Length')
plt.xticks(rotation=45)
plt.show()
```

# Churn by Gender

```python
sns.countplot(x='Gender', hue='Churn', data=df)
plt.title('Churn Distribution by Gender')
plt.show()
```

# Churn by Subscription Type


```python
sns.countplot(x='Subscription Type', hue='Churn', data=df)
plt.title('Churn Distribution by Subscription Type')
plt.xticks(rotation=45)
plt.show()
```

# Churn by Age

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Age', hue='Churn', kde=True, multiple='stack')
plt.title('Churn Distribution by Age')
plt.show()
```

# Correlation matrix


```python
corr_matrix = df.corr(numeric_only=True)
```

# Correlation heatmap

```python
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# Distribution of Tenure by Churn

```python
sns.histplot(data=df, x='Tenure', hue='Churn', kde=True, multiple='stack')
plt.title('Tenure Distribution by Churn')
plt.show()
```

# Distribution of Usage Frequency by Churn

```python
sns.histplot(data=df, x='Usage Frequency', hue='Churn', kde=True, multiple='stack')
plt.title('Usage Frequency by Churn')
plt.show()
```

# Distribution of Total Spend by Churn

```python
sns.histplot(data=df, x='Total Spend', hue='Churn', kde=True, multiple='stack')
plt.title('Total Spend Distribution by Churn')
plt.show()
```

# Boxplot of Total Spend by Churn

```python
sns.boxplot(x='Churn', y='Total Spend', data=df)
plt.title('Total Spend vs. Churn')
plt.show()
```

# Boxplot of Payment Delay by Churn


```python
sns.boxplot(x='Churn', y='Payment Delay', data=df)
plt.title('Payment Delay vs. Churn')
plt.show()
```

# Boxplot of Tenure by Churn

```python
sns.boxplot(x='Churn', y='Tenure', data=df)
plt.title('Tenure vs. Churn')
plt.show()
```
